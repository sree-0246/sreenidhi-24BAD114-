SCENARIO 1: E-commerce Sales Data
Analyze a sample e-commerce dataset to understand product categories, sales trends, and missing information.
Dataset (Kaggle â€“ Public):  
https://www.kaggle.com/datasets/carrie1/ecommerce-data

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("SREENIDHI - 24BAD114\n")

df = pd.read_csv("data.csv", encoding="ISO-8859-1")

# Inspect dataset
print("----- HEAD -----")
print(df.head())

print("\n----- TAIL -----")
print(df.tail())

print("\n----- INFO -----")
df.info()

print("\n----- DESCRIBE -----")
print(df.describe())

# Check missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# Create Total Sales column
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Calculate total sales per product
sales_per_product = (
    df.groupby('Description')['TotalSales']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Bar chart - Top 10 products by sales
plt.figure()
sales_per_product.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()

# Line chart - Sales trend of top 10 products
plt.figure()
sales_per_product.plot(kind='line', marker='o')
plt.title("Sales Trend of Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

SCENARIO 2: Hospital Patient Records
Examine a patient dataset to identify missing health metrics and visualize distributions.
Dataset (Kaggle â€“ Public):
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("SREENIDHI - 24BAD114\n")

df = pd.read_csv("diabetes.csv")

# Explore dataset structure
print("----- HEAD -----")
print(df.head())

print("\n----- INFO -----")
df.info()

print("\n----- DESCRIBE -----")
print(df.describe())

# Check missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# Note: In this dataset, zero values indicate missing health metrics
columns_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
print("\n----- ZERO VALUES (TREATED AS MISSING) -----")
print((df[columns_with_zero] == 0).sum())

# Histogram: Glucose levels
plt.figure()
plt.hist(df['Glucose'], bins=20)
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.show()

# Histogram: Age distribution
plt.figure()
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Boxplot: Glucose levels
plt.figure()
plt.boxplot(df['Glucose'])
plt.title("Boxplot of Glucose Levels")
plt.ylabel("Glucose Level")
plt.show()

# Boxplot: Age
plt.figure()
plt.boxplot(df['Age'])
plt.title("Boxplot of Age Distribution")
plt.ylabel("Age")
plt.show()

# Identify patterns in health metrics
print("\n----- HEALTH METRIC PATTERNS -----")
print(df.groupby('Outcome')[['Glucose', 'BMI', 'Age']].mean())
```

SCENARIO 3: Housing Dataset
Investigate a housing dataset for missing features and relationships between house attributes.
Dataset (Kaggle â€“ Public):
https://www.kaggle.com/datasets/harishkumardatalab/housing-price-prediction

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("SREENIDHI - 24BAD114\n")

df = pd.read_csv("Housing.csv")

# Inspect dataset
print("----- COLUMNS -----")
print(df.columns)

print("\n----- HEAD -----")
print(df.head())

print("\n----- INFO -----")
df.info()

# Missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

print("\n----- DESCRIBE -----")
print(df.describe())

# Scatter plot: Area vs Price
plt.figure()
plt.scatter(df['area'], df['price'])
plt.title("Area vs House Price")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()

# Scatter plot: Bedrooms vs Price
plt.figure()
plt.scatter(df['bedrooms'], df['price'])
plt.title("Bedrooms vs House Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()

# Correlation heatmap using matplotlib (NO seaborn)
corr = df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap of Housing Dataset")
plt.show()
```

SCENARIO 4: Banking Customer Data
Analyze banking data to detect missing account information and understand customer demographics.
Dataset (Kaggle â€“ Public):https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("SREENIDHI - 24BAD114\n")

df = pd.read_csv("marketing_campaign.csv", sep='\t')

# Examine structure
print("----- HEAD -----")
print(df.head())

print("\n----- INFO -----")
df.info()

print("\n----- COLUMN TYPES -----")
print(df.dtypes)

# Check missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# Create Age column from Year_Birth
df['Age'] = 2024 - df['Year_Birth']

# ---------------- VISUALIZATIONS ----------------

# Bar plot: Age distribution (binned)
plt.figure()
df['Age'].plot(kind='hist', bins=10)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()

# Box plot: Income
plt.figure()
plt.boxplot(df['Income'].dropna())
plt.title("Income Distribution of Customers")
plt.ylabel("Income")
plt.show()

# Spending pattern: Total spending
df['Total_Spending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)

# Box plot: Spending
plt.figure()
plt.boxplot(df['Total_Spending'])
plt.title("Customer Spending Pattern")
plt.ylabel("Total Spending")
plt.show()
```
